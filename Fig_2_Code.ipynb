```python
#initial intensity not equal to 0, constant source function
import numpy as np
import matplotlib.pyplot as plt
c= 3*10**10 #velocity of light cm
h=6.62*10**(-27) #planck constant erg s
k=1.38*10**(-16) #boltzmann constant erg degree^{-1}
T=6000 #temperature of sun K
s=1 #path length
lambda_=483.3*10**(-7)
planck_func=2*h*c**2/lambda_**5*(1/(np.exp(h*c/(lambda_*k*T))-1))
S_lambda=planck_func
tau_lambda=np.linspace(0,5,1000)
I_lambdasmall1=0.5*S_lambda+tau_lambda*(S_lambda-0.5*S_lambda)
I_lambda1=0.5*S_lambda*np.exp(-tau_lambda)+S_lambda*(1-np.exp(-tau_lambda))
I_lambdasmall2=1.5*S_lambda+tau_lambda*(S_lambda-1.5*S_lambda)
I_lambda2=1.5*S_lambda*np.exp(-tau_lambda)+S_lambda*(1-np.exp(-tau_lambda))
scale=1*10**(14)
plt.figure(figsize=(6.8,5.3))
plt.plot(tau_lambda,I_lambdasmall1/scale, linestyle='--',label=r"$\tau_\lambda\ll1$ approximation, $I_{\lambda0}<S_0$")
plt.plot(tau_lambda,I_lambdasmall2/scale,linestyle='--',label=r"$\tau_\lambda\ll1$ approximation, $I_{\lambda0}>S_0$")
plt.plot(tau_lambda,I_lambda1/scale,label=r"Exact solution, $I_{\lambda0}< S_\lambda$",color='blue')
plt.plot(tau_lambda,I_lambda2/scale, label=r"Exact solution, $I_{\lambda0}> S_\lambda$",color='green')
plt.axhline(planck_func/scale, label=r"Source function $S_\lambda$", color='m')
plt.xlabel(r'Optical depth $\tau_\lambda$',fontsize=12)
plt.ylabel(r'Intensity $I_\lambda$ ($\times 10^{14}$ erg s$^{-1}$ cm$^{-2}$ sr$^{-1}$ cm$^{-1}$) ', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=10, loc='upper left')
plt.grid()
plt.savefig("initial intensity not equal to 0.pdf")
plt.show()
```
