```python
#source function
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
c= 3*10**10 #velocity of light cm
h=6.62*10**(-27) #planck constant erg s
k=1.38*10**(-16) #boltzmann constant erg degree^{-1}
T=6000 #temperature of sun K
s=1 #path length
lambda_ = np.linspace(1*10**(-5),1*10**(-4),500)
tau_lambda=np.linspace(0,5,500)
scale=1*10**14
tau_lambda, lambda_= np.meshgrid(tau_lambda, lambda_)
S_lambda=2*h*c**2/lambda_**5*(1/(np.exp(h*c/(lambda_*k*T))-1))
plt.pcolormesh(tau_lambda,lambda_*10**7,S_lambda/scale,shading='auto',cmap='plasma')
plt.xlabel(r"Optical Depth $\tau_\lambda$", fontsize=12)
plt.ylabel(r"Wavelength $\lambda$ (nm)",fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.clim(0,4.5)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=12)
cbar.set_label(r"Source Function $S_\lambda$ ($\times 10^{14}$ erg s$^{-1}$ cm$^{-2}$ sr$^{-1}$ cm$^{-1}$)", fontsize=12)
plt.contour(tau_lambda, lambda_*10**7, S_lambda/scale, colors='black',linewidths=0.5)
plt.tight_layout()
plt.savefig("source function.pdf")
plt.show()
```
