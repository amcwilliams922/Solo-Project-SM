```python
#initial intensity equals 0, constant source function
import numpy as np
import matplotlib.pyplot as plt
c= 3*10**10 #velocity of light cm
h=6.62*10**(-27) #planck constant erg s
k=1.38*10**(-16) #boltzmann constant erg degree^{-1}
T=6000 #temperature of sun K
s=1 #path length
lambda_=483.3*10**(-7)
planck_func=2*h*c**2/lambda_**5*(1/(np.exp(h*c/(lambda_*k*T))-1))
S_lambda=planck_func
tau_lambda=np.linspace(0,5,1000)
I_lambdasmall=planck_func*tau_lambda
I_lambda=S_lambda*(1-np.exp(-tau_lambda))
scale=1*10**(14)
plt.plot(tau_lambda, I_lambdasmall/scale, label=r"$\tau_\lambda\ll1$ approximation", linestyle='--')
plt.plot(tau_lambda, I_lambda/scale ,label=r"Exact solution", color='blue')
plt.axhline(planck_func/scale, color='m', label=r"Source function $S_\lambda$")
plt.xlabel(r'Optical depth $\tau_\lambda$', fontsize=12)
plt.ylabel(r'Intensity $I_\lambda$ ($\times 10^{14}$ erg s$^{-1}$ cm$^{-2}$ sr$^{-1}$ cm$^{-1}$) ', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.legend(fontsize=12)
plt.savefig("initial intensity=0.pdf")
plt.show()
```
