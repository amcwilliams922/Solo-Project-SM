```python
#initial intensity equals 1.5*S_lambda, difference in the source funsction and initial intensity
import numpy as np
import matplotlib.pyplot as plt
c= 3*10**10 #velocity of light cm
h=6.62*10**(-27) #planck constant erg s
k=1.38*10**(-16) #boltzmann constant erg degree^{-1}
T=6000 #temperature of sun K
s=1 #path length
tau_lambda=np.linspace(0,5,1000)
lambda_ = np.linspace(1*10**(-5),1*10**(-4),1000)
planck_func=2*h*c**2/lambda_**5*(1/(np.exp(h*c/(lambda_*k*T))-1))
lambda0=1.5*planck_func
I_lambda=1.5*planck_func*np.exp(-tau_lambda)+planck_func*(1-np.exp(-tau_lambda))
scale=1*10**14

plt.plot(lambda_*10**7, abs(planck_func-lambda0)/scale)
plt.xlabel(r'Wavelength $\lambda$ (nm)', fontsize=12)
plt.ylabel(r"$|S_\lambda-I_{\lambda0}|$ ($\times 10^{-10}$ erg s$^{-1}$ cm$^{-2}$ sr$^{-1}$ cm$^{-1}$)", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.savefig('difference.pdf')
plt.show()
```
