```python
#initial intensity equals 1.5*S_lambda, shorter wavelength
import numpy as np
import matplotlib.pyplot as plt
c= 3*10**10 #velocity of light cm
h=6.62*10**(-27) #planck constant erg s
k=1.38*10**(-16) #boltzmann constant erg degree^{-1}
T=6000 #temperature of sun K
s=1 #path length
lambda1=100
planck_func1=2*h*c**2/lambda1**5*(1/(np.exp(h*c/(lambda1*k*T))-1))
tau_lambda=np.linspace(0,5,1000)
I_lambda1=1.5*planck_func1*np.exp(-tau_lambda)+planck_func1*(1-np.exp(-tau_lambda))
scale=1*10**10

plt.plot(tau_lambda, I_lambda1*scale ,label=r"Exact solution")
plt.axhline(planck_func1*scale, color='magenta', label=r"Source function $S_\lambda$")
plt.xlabel(r'Optical depth $\tau_\lambda$', fontsize=12)
plt.ylabel(r"Intensity $I_\lambda$ ($\times 10^{-10}$ erg s$^{-1}$ cm$^{-2}$ sr$^{-1}$ cm$^{-1}$)", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
